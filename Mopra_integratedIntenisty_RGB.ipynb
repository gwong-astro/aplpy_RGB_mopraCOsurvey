```python
%matplotlib inline
import aplpy
import montage_wrapper as montage
```

```python
#create the RGB cube
aplpy.make_rgb_cube(['G300-310-12CO.fits', 'G300-310-13CO.fits', 'G300-310-C18O.fits'], 'G300-310_colour.fits')
aplpy.make_rgb_cube(['G310-320-12CO.fits', 'G310-320-13CO.fits', 'G310-320-C18O.fits'], 'G310-320_colour.fits')
aplpy.make_rgb_cube(['G320-330-12CO.fits', 'G320-330-13CO.fits', 'G320-330-C18O.fits'], 'G320-330_colour.fits')
aplpy.make_rgb_cube(['G330-340-12CO.fits', 'G330-340-13CO.fits', 'G330-340-C18O.fits'], 'G330-340_colour.fits')
aplpy.make_rgb_cube(['G340-350-12CO.fits', 'G340-350-13CO.fits', 'G340-350-C18O.fits'], 'G340-350_colour.fits')

# Make an RGB image, you can change the scale
aplpy.make_rgb_image('G300-310_colour.fits', 'G300-310_colour.png')
aplpy.make_rgb_image('G310-320_colour.fits', 'G310-320_colour.png')
aplpy.make_rgb_image('G320-330_colour.fits', 'G320-330_colour.png')
aplpy.make_rgb_image('G330-340_colour.fits', 'G330-340_colour.png')
aplpy.make_rgb_image('G340-350_colour.fits', 'G340-350_colour.png')
# ,vmin_r=0., pmax_g=0.
```

```python
# Plot the RGB image using the 2d image to indicate the projection
img = aplpy.FITSFigure('G300-310-12CO.fits')
img.set_tick_color('white')
img.set_tick_labels_style('latex')
img.set_tick_labels_font(family='serif')
img.set_axis_labels(xlabel='Galatic longitude', ylabel='Galatic latitude')
img.set_xaxis_coord_type('longitude')
img.set_yaxis_coord_type('latitude')
img.set_tick_labels_format(xformat='dd.d',yformat='dd.d')
img.set_axis_labels_font(family='serif')
img.show_rgb('G300-310_colour.png')
img.save('G300-310_colour_test.pdf')

# Plot the RGB image using the 2d image to indicate the projection
img = aplpy.FITSFigure('G310-320-12CO.fits')
img.set_tick_color('white')
img.set_tick_labels_style('latex')
img.set_tick_labels_font(family='serif')
img.set_axis_labels(xlabel='Galatic longitude', ylabel='Galatic latitude')
img.set_xaxis_coord_type('longitude')
img.set_yaxis_coord_type('latitude')
img.set_tick_labels_format(xformat='dd.d',yformat='dd.d')
img.set_axis_labels_font(family='serif')
img.show_rgb('G310-320_colour.png')
img.save('G310-320_colour_test.pdf')

# Plot the RGB image using the 2d image to indicate the projection
img = aplpy.FITSFigure('G320-330-12CO.fits')
img.set_tick_color('white')
img.set_tick_labels_style('latex')
img.set_tick_labels_font(family='serif')
img.set_axis_labels(xlabel='Galatic longitude', ylabel='Galatic latitude')
img.set_xaxis_coord_type('longitude')
img.set_yaxis_coord_type('latitude')
img.set_tick_labels_format(xformat='dd.d',yformat='dd.d')
img.set_axis_labels_font(family='serif')
img.show_rgb('G320-330_colour.png')
img.save('G320-330_colour_test.pdf')


# Plot the RGB image using the 2d image to indicate the projection
img = aplpy.FITSFigure('G330-340-12CO.fits')
img.set_tick_color('white')
img.set_tick_labels_style('latex')
img.set_tick_labels_font(family='serif')
img.set_axis_labels(xlabel='Galatic longitude', ylabel='Galatic latitude')
img.set_xaxis_coord_type('longitude')
img.set_yaxis_coord_type('latitude')
img.set_tick_labels_format(xformat='dd.d',yformat='dd.d')
img.set_axis_labels_font(family='serif')
img.show_rgb('G330-340_colour.png')
img.save('G330-340_colour_test.pdf')

# Plot the RGB image using the 2d image to indicate the projection
img = aplpy.FITSFigure('G340-350-12CO.fits')
img.set_tick_color('white')
img.set_tick_labels_style('latex')
img.set_tick_labels_font(family='serif')
img.set_axis_labels(xlabel='Galatic longitude', ylabel='Galatic latitude')
img.set_xaxis_coord_type('longitude')
img.set_yaxis_coord_type('latitude')
img.set_tick_labels_format(xformat='dd.d',yformat='dd.d')
img.set_axis_labels_font(family='serif')
img.show_rgb('G340-350_colour.png')
img.save('G340-350_colour_test.pdf')

```
